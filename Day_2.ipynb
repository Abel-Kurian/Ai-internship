```python
df = pd.read_csv("/content/preprocessing_practice_dataset.csv")
```

```python
df['Age_mean_inputed']=df['Age'].fillna(df['Age'].mean())
df['Age_median_inputed']=df['Age'].fillna(df['Age'].median())

df['Department_mean_inputed']=df['Department'].fillna(df['Department'].mode())


df['Education_mean_inputed']=df['Education'].fillna(df['Education'].mode())


df['Monthly_Salary_mean_inputed']=df['Age'].fillna(df['Age'].mean())
df['Monthly_Salary_median_inputed']=df['Age'].fillna(df['Age'].median())

df['City_mean_inputed']=df['Age'].fillna(df['Age'].mode())

```

```python
df
```

```python


sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Department',y='Monthly_Salary',palette='pastel')


plt.title('Boxplot of Age by Passenger Class',fontsize=12)
plt.xlabel('Embarked',fontsize=12)
plt.ylabel('Fare',fontsize=12)

plt.show()
```

```python
sns.scatterplot(data=df,x='Department',y='Monthly_Salary',palette='pastel')
plt.title('Scatter of Age by Passenger Class',fontsize=12)
plt.xlabel('Embarked',fontsize=12)
plt.ylabel('Fare',fontsize=12)

plt.show()
```

```python
sns.lineplot(data=df,x='Department',y='Monthly_Salary',palette='pastel')
plt.title('Lineplot of Age by Passenger Class',fontsize=12)
plt.xlabel('Embarked',fontsize=12)
plt.ylabel('Fare',fontsize=12)

plt.show()
```

```python
sns.histplot(data=df,x='Department',y='Monthly_Salary',palette='pastel')
plt.title('Histogram of Age by Passenger Class',fontsize=12)
plt.xlabel('Embarked',fontsize=12)
plt.ylabel('Fare',fontsize=12)

plt.show()
```
